# Outlier Detection

```python
data_c2['Ozone'].hist()
```

```python
plt.boxplot(data_c2['Ozone'].dropna())
```

```python
outliers=[]

def det_outliers(data):
  mean=np.mean(data)
  std=np.std(data)
  thresold=3
  for i in data:
    z_score=(i-mean)/std
    if np.abs(z_score)>thresold:
      outliers.append(i)

  return outliers
```

```python
det_outliers(data['Ozone'].dropna())
```

```python
data_c3.mean(), data_c3.std()
```

```python
outliers_iqr=[]
def det_otliers_iqr(data):
  q1, q3= np.percentile(data, [25,75])
  iqr=q3-q1
  lower_fence= q1-1.5*iqr
  upper_fence= q3+1.5*iqr
  for i in data:
    if i<lower_fence:
      outliers_iqr.append(i)
    elif i> upper_fence:
      outliers_iqr.append(i)
  return outliers_iqr
```

```python
det_otliers_iqr(data['Ozone'].dropna())
```
